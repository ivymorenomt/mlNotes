```python
import numpy as np
import matplotlib.pyplot as plt
```

```python

pageSpeeds = np.random.normal(3.0, 1.0, 1000)
purchaseAmount = 100 - (pageSpeeds + np.random.normal(0, 0.1, 1000)) * 3

plt.scatter(pageSpeeds, purchaseAmount)
```

## As we only have two features, we can keep it simple and just use scipy.state.lingress

```python
from scipy import stats

slope, intercept, r_value, p_value, std_err= stats.linregress(pageSpeeds, purchaseAmount)
```

Not Surprisingly our R-squared value shows a really good fit:

```python
r_value **2
```

Let's use slope and intercept we got from regression to plot predicted values vs observed 

```python
def predict(x):
    return slope * x + intercept # this is based on linregress we took
fitLine = predict(pageSpeeds)

plt.scatter(pageSpeeds, purchaseAmount)
plt.plot(pageSpeeds, fitLine, c='r')
plt.show()
```
