```python
import pandas as pd
from sklearn.datasets import load_breast_cancer
```

### Data set
It is built in from scikit learn, breast cancer dataset is a classification dataset.
It consists of calculations of measurements done on a tumor that was collected or mass that is collected. We can use this to implement decision trees

Target Label - learning if a line is a tumor or benign

```python
data = load_breast_cancer()
dataset = pd.DataFrame(data=data['data'], columns=data['feature_names'])
dataset
```

## Train/Test

We need to separate training data and testing dataset
X is the features, or all the columns to determine if the datapoint is benign or malignant
y is the target value, so it would be 0 or 1 or good or bad.

- Test Size = how much are you going to set aside for testing?
- Training data = we are giving the samples and the answers so the model can train or learn from
- Testing data = we are giving the columns only to create the predictions itself so we could test if the model is performing or not.

```python
from sklearn.model_selection import train_test_split
X = dataset.copy()
y = data['target']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33)
```

```python
from sklearn.tree import DecisionTreeClassifier
clf = DecisionTreeClassifier(ccp_alpha=0.01)
clf = clf.fit(X_train, y_train) # Classifier fit You give the x values and y values here, then this will create your decision tree
```

```python
clf.get_params() # we are using the default parameters
```

```python
# Predict
predictions = clf.predict(X_test)
predictions
```

The array represents the predictions, whether it is benign or malignant. Another way of predictiing is using predict_proba. It will give you the probability for each class. For the first datapoint, it would be 0, 1 out of 1 chance.
* Currently we don't have a stopping criteria that is why our tree grew. (modified the tree to max_depth=4)

Now we have the predictions and predictions for each of instances, but how do we know if this is good or not? We have to compare by using a performance metrics.

The confusion matrix Predicted Positive/Actual Positive(64) - it means 0 was predicted 69 times -  which is positively predicted, then Predicted Negative/Actual Negative(110) - it means 1 was predicted 110 times - and 5, 9 are wrongly classified predictions

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test, predictions)
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test, predictions, labels=[0,1])
```

```python
from sklearn.metrics import precision_score
precision_score(y_test, predictions)
```

```python
from sklearn.metrics import recall_score
recall_score(y_test, predictions)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, predictions, target_names=['malignant', 'benign']))
```

## Decision Tree Parameters

These are the stopping criteria:
- 'max_depth' - how deep the tree is (we would not know the real value, but we have to try which works best)
- 'min_samples_split'
- 'min_samples_leaf'
- 'min_weight_fraction_leaf
- 'max_leaf_nodes'
- 'min_impurity_decrease'
- min_impurity_split

Approaches of Decision tree
We use CART and ID3 algorithm, and we use different features to split the tree:
- criterion - gini(default), entropy
- splitter - best(default), random 
- max_features - int, None(default) - everytime you want to make a split, only use (20) of them and then it will compare the entropy or gini information to each other and select the best one. It will only select the best within that 20
- random_state - int/float, None(default) - randomization of columns; you will get the same tree over and over if you pick random_state to 5 or 1.

- class_weight
